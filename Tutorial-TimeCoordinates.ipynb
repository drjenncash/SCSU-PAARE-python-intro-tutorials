## Astropy Units

`astropy.units` handles defining, converting between, and performing arithmetic with physical quantities, such as meters, seconds, Hz, etc. It also handles logarithmic units such as magnitude and decibel.

The available units include the standard SI, cgs, and astronophysical units. You can also enable imperial units and define custom units. 



The basic units are:
Length (meter)
,
Time (second)
,
Mass (kilogram)
,
Current (ampere)
,
Temperature (Kelvin)
,
Angular distance (radian)
,
Solid angle (steradian)
,
Luminous intensity (candela)
,
Stellar magnitude (mag)
,
Amount of substance (mole)
,
Photon count (photon)

### Math with units

When using quantities that have units, the math must follow the units rules. 

* When adding or subtracting, the values must have compatible units
    * if the units are the same, it gives the sum in that unit
    * if the units are compatible, the result will be the in the units 

```python

print('a+b gives the result in meters, since a was in meters', a + b)
print('b+a gives the result in kilometers, since b was in meters', b+a)
```

* When multiplying or dividing units, you get a composite unit
    * common units are converted to the unit squared or reduced version automatically
    * you may want to convert it into a specific quantity or reduce it manually if they aren't the same
  
    

```python
print('a*a gives', a*a)
print('b*b gives', b*b)
print('a*b gives', a*b)
print('a*b gives', (a*b).to(u.m**2))
print('a/a gives', a/a)
print('b/b gives', b/b)
print('a/b gives', a/b)
print('reducing a/b gives', (a/b).decompose())
```

```python
dist = 75 *u.km
time = 35*u.min
print(dist/time)
print((dist/time).to(u.m/u.s))
```

```python
speed = 75 *u.km/u.h
time = 32 *u.min
dist = (speed*time).decompose()
print(dist)
```

## Astropy Constants

A number of important physical constants are included in the `astropy.constants` module.

You can import the entire set of constants using 
`from astropy import constants as const`

Or import just the one or two you may need for a specific piece of code

`from astropy.constants import M_sun`
`from astropy.constants import sigma_sb`

The structure of your import statement will determine how you call the function.

```python
from astropy import constants as const
print(const.G)
print()
print(const.m_p)
```

```python
from astropy.constants import M_sun
print(M_sun)
```

The constants have units and those units will be preserved in any calculation that uses them.

```python
print(3*M_sun)
```

## Astropy Time


In astropy.time a “time” is a single instant of time which is independent of the way the time is represented (the “format”) and the time “scale” which specifies the offset and scaling relation of the unit of time. There is no distinction made between a “date” and a “time” since both concepts (as loosely defined in common usage) are just different representations of a moment in time.
